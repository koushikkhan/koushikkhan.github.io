# Introduction

In our high school mathematics, we have learnt *Differential Calculus*, which has introduced the term 'differentiation' or 'derivative' of a function with respect to a variable. This derivative has played a significant role in today's deep learning algorithms. 

We all are familiar with the concept of derivative and hopefully, have solved a lot of exercises. I this post, I will revisit the core idea of computing derivativea and also try to develop the idea that is being used behind the learning process of a neural network.

# Derivative of a function

Let us take a simple function $f(.)$ that depends on the input $x$. The derivative of $f(x)$ w.r.t $x$ is defined as follows:

$$
\frac{df}{dx} = \lim_{h \to 0} \frac{f(x+h) - f(x)}{(x+h) - x}
$$

The animation below, made with [geogebra](https://www.geogebra.org), will make the definition easier to understand.

As you understand, the ratio $\frac{f(x+h) - f(x)}{(x+h) - x}$ is nothing but the *tangent* or $tan()$ of the angle $\angle{QPP'}$ from the triangle $\triangle{QPP'}$.

Theoratically, when $h$ is very close to zero, the ratio gives you the value of the tangent of the curve at the point $(x,0)$.

Therefore, the overall definition of derivative of a function, geometrically refers to the tangent value of the curve $y = f(x)$ at the point $x$.

The tangent of a curve at some point $x$ is important as it gives us an idea about the nature of the curve locally. Geometrically, if the tangent line makes an acute angle with the $X-Axis$ (which is the case here), then it denotes a positive change on the output and when it makes an obtuse angle, it denotes a negative change on the output. If the tangent line horizontal to the $X-Axis$, then there is no change on the output.

Similar information can also be obtained by evaluating the expression of derivative at some point $x$. If the limiting value of the ratio is positive, then obviously, there is a positive change on the output otherwise the change is negative and having a value as zero indicates there is no change on the output.

Note that, till now, we have discussed only on a scalar valued function with a scalar valued input. We will now try to get some ideas to calculate derivative when either of the function and input or both of them are vector valued.

# Derivative of vector valued functions

## Functions of type $f: \mathbb{R}^n \to \mathbb{R}$

Such functions take vector (s) as input (s) and return a scalar. Loss functions like **MSE** or **Cross Entropy** can be couple of examples.

In this case, derivative of $f(\vec{x})$ will also be a vector of the same dimension, as given below:

$$
\frac{df}{d\vec{x}} = \left(\frac{df}{dx_1}, \ldots, \frac{df}{dx_n}\right)
$$


## Functions of type $f: \mathbb{R}^n \to \mathbb{R}^m$

Such functions take vector (s) as input (s) and return another vector with different dimension as output. 

A simple example can be a function $f: \mathbb{R}^2 \to \mathbb{R}^3$:

+ input: $\vec{x} = \left(x_1, x_2\right)$ and a matrix of parameters $\mathbf{W}_{3 \times 2}$
+ output: $\vec{y} = \mathbf{f(\vec{x})} = \left(w_{11}x_1+w_{12}x_2, w_{21}x_1+w_{22}x_2, w_{31}x_1+w_{32}x_2)\right)$

In this case, derivative of $\mathbf{f(\vec{x})}$ will be a matrix of dimension $m \times n$, as given below:

\begin{align}
    \frac{d\vec{y}}{d\vec{x}} &= \left(\frac{d\vec{y}}{dx_1}, \ldots, \frac{d\vec{y}}{dx_n}\right) \\
    &= \begin{pmatrix}
            \frac{dy_1}{dx_1} & \frac{dy_1}{dx_2} & \ldots & \frac{dy_1}{dx_n} \\
            \frac{dy_2}{dx_1} & \frac{dy_2}{dx_2} & \ldots & \frac{dy_2}{dx_n} \\
            \vdots & \vdots & \ddots & \vdots \\
            \frac{dy_m}{dx_1} & \frac{dy_m}{dx_2} & \ldots & \frac{dy_m}{dx_n} \\
        \end{pmatrix}
\end{align}

This matrix is often referred to as **Jacobian** of $\vec{y}$ with respect to $\vec{x}$.

# Computational graph

A *computational graph* is a directed acyclic graph that keeps track of the sequence of operations performed during the run time of an algorithm.

Let us take an example to understand it.


```{mermaid}
%%| fig-width: 4.5
graph LR
  a((a = 3)) --> c((c = a+1))
  a((a = 3)) --> d((d = a+b))
  b((b = 2)) --> d((d = a+b))
  c((c = a+1)) --> e((e = c * d))
  d((d = a+b)) --> e((e = c * d)) 
```


The above diagram is typically referred to as a computational graph, where both **a** and **b** are inputs to the graph and **e** is the output of the graph.

Each of the boxes in the graph is called a *node* where a mathematical operation is performed.

::: {.border font-size="22px" font-style="italic"}
This content can be styled with a border.
:::


```{block, type="quarto"}
.classDef my-class color:#f00, font-weight:bold;
```
